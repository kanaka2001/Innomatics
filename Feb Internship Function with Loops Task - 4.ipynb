Problem 1: Mobile Recharge Validation System

```python
# Function to validate recharge
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Check minimum amount
    if amount < 50:
        return "Amount must be at least ₹50"

    # Check if plan is valid
    if amount not in valid_plans:
        return "Invalid recharge plan"

    return "Recharge Successful"


# Retry using while loop
while True:
    amt = int(input("Enter recharge amount: ")) 
    result = validate_recharge(amt)
    print(result)

    if "Successful" in result:
        break
```

Problem 2: Inventory Reorder Alert System

```python
# Function to check stock
def check_inventory(products):
    for item, qty in products.items():
        if qty < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")


# Sample data
products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12
}

check_inventory(products)
```

Problem 3: Student Result Processing System

```python
# Function to process result
def process_result(marks):
    total = 0

    # Calculate total using loop
    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return f"Average = {avg} → Pass"
    else:
        return f"Average = {avg} → Fail"


# Example marks
marks = [60, 90, 70, 40]

print(process_result(marks))
```

Problem 4: Cab Fare Estimator with Retry Option

```python
# Function to calculate fare
def calculate_fare(km, peak):
    base = 50
    fare = base + (km * 12)

    if peak == "yes":
        fare += fare * 0.25  # 25% extra

    return fare


# Retry loop
while True:
    distance = float(input("Enter distance (km): "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total = calculate_fare(distance, peak_hour)
    print("Total Fare = ₹", total)

    again = input("Retry? (yes/no): ")
    if again.lower() != "yes":
        break
```

Problem 5: Employee Attendance Eligibility checker

```python
# Function to check eligibility
def attendance_check(attendance):
    present = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance = {percentage}% → Eligible"
    else:
        return f"Attendance = {percentage}% → Not Eligible"


# Example attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P"]

print(attendance_check(attendance))
```

Problem 6: Password Strength Checker 

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Conditions
    if len(password) < 8:
        return "Password too short"

    if not has_digit:
        return "Must contain a digit"

    if not has_special:
        return "Must contain special character (@#$)"

    return "Strong Password"


# Test password
pwd = input("Enter password: ")
print(check_password(pwd))
```
