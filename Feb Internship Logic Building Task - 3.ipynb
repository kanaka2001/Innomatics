```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
max_score = max(employees.values())

# Find all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == max_score]

# Display the result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

# Problem 2: Search Query Keyword Analysis

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency of each word
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Keep only words that appear more than once
repeated_words = {word: count for word, count in frequency.items() if count > 1}

# Display the result
print(repeated_words)
```

# Problem 3: Sensor Data Validation

```python


sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through readings with index (hour)
for index, value in enumerate(sensor_readings):
    # Check if the reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((index, value))

# Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem 4: Email Domain Usage Analysis

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count users per domain
for email in emails:
    domain = email.split("@")[1]  # Extract domain part
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

# Problem 5: Sales Spike Detection


```python

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Define threshold (30% above average)
threshold = average_sales * 1.3

# Find and display days with sales above threshold
for index, value in enumerate(sales, start=1):  # start=1 for day numbers
    if value > threshold:
        print(f"Day {index}: {value}")
```

# Problem 6: Duplicate User ID Detection

```python


user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences of each user ID
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display only duplicates (count > 1)
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
