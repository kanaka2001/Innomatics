Problem 4: Social Media – Duplicate Account Detection

```python
# Function to identify and list duplicate usernames from a provided list
def DuplicateAccountDetection(listofusernames):
    # 'seen' tracks unique names; 'duplicates' stores names found more than once
    seen = set()
    duplicates = set()

    for name in listofusernames:
        # If the name is already in 'seen', it's a duplicate
        if name in seen:
            # Add to the duplicates set (sets automatically handle multiple repeats)
            duplicates.add(name)
        else:
            # Otherwise, record it as a seen name for future comparisons
            seen.add(name)

    # Check if the duplicates set contains any items
    if duplicates:
        # Join the set of duplicates into a single string separated by commas
        print(f"Duplicate Accounts Found: Yes")
        print(f"Flagged Usernames: {', '.join(duplicates)}")
    else:
        # Message if all usernames in the list are unique
        print("Duplicate Accounts Found: No")

# Sample data containing intentional duplicates (loki and thor)
account_list = ["loki", "thor", "ironman", "loki", "nature_lover", "loki", "thor"]

# Run the detection algorithm
DuplicateAccountDetection(account_list)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter and display crops that exceed a specific price threshold
def PremiumCropPriceFilter(crop_prices, cap_rate):
        # Use list comprehension to create a new list of prices higher than the cap_rate
        # This iterates through crop_prices and keeps only those that meet the condition
        listofpremiumcrops = [price for price in crop_prices if price > cap_rate]
        
        # Print the resulting list of premium prices
        print(f"Premium Crop: {listofpremiumcrops}")

# Initial list of crop prices to be evaluated
crop_prices = [100, 150, 80, 200, 50]

# Prompt the user to enter a numeric threshold for filtering
cap_rate = int(input("Enter the price cap to filter premium crops: "))

# Call the function with the prices and the user-defined cap
PremiumCropPriceFilter(crop_prices, cap_rate)
```

Problem 7: System Monitoring – Application Health Checker

```python
# Function to monitor system health based on error counts
def apphealthmonitor(healthreadings):
    # Loop through each individual health reading (error count)
    for healthreading in healthreadings:
        # Use structural pattern matching to categorize system status
        match healthreading:
            # Case for perfectly healthy systems (zero errors)
            case healthreading if healthreading == 0:
                status = "Healthy"
            # Case for systems with a low volume of errors (1 to 5)
            case healthreading if healthreading <= 5:
                status = "Minor Issues"          
            # Case for systems with high error counts (above 5)
            case healthreading if healthreading > 5:
                status = "Critical Issues"
        
        # Display the current error count and the evaluated status
        print(f"Error Count : {healthreading}\n System Status: {status}\n")

# List of error readings from various app components
healthreadings = [0, 3, 6, 2, 8]    

# Run the monitor to process the system health data
apphealthmonitor(healthreadings)
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def withdrawalmonitor():
    total_withdrawn = 0
    while True:
        user_choice = input("Enter amount to withdraw or 'q' to quit: ")
        
        if user_choice.lower() == 'q':
            break
            
        amount = int(user_choice)
        status = "Rejected" if amount >= 50000 else "Approved"
        
        # Only add to total if the bank actually approves it!
        if status == "Approved":
            total_withdrawn += amount 
            print(f"Transaction Amount: {amount}\nStatus: {status}")
            print(f"Total Withdrawn Amount: {total_withdrawn}\n")
        elif status == "Rejected":
            print(f"Transaction Rejected: Amount exceeds the withdrawal limit.\n the amount you can withdraw now is {50000- total_withdrawn}")

withdrawalmonitor()
```
