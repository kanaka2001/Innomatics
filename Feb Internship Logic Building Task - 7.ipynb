Problem 1: Smart Light Controller

```python
#class to control lights
class control:
  def __init__(self,light_name):
    self.light_name=light_name
    self.is_on=False
  #functionto turn on the light
  def on(self):
    self.is_on=True
  def off(self):
  #function to turn off the light
    self.is_on=False
  #function to display status
  def display(self):
    if self.is_on:
      print(f"{self.light_name} is ON")
    else: 
      print(f"{self.light_name} is OFF")
#sample input
light=control("Bedroom Light")
light.on()
light.display()
```

Problem 2: Employee ID Card System

```python
#class to store employee name, ID, and department and displays ID card details
class EmployeeID:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department
    #function to display
  def display(self):
    print("Employee ID Card")
    print(f"Employee Name:{self.name}")
    print(f"Employee ID:{self.id}")
    print(f"Department:{self.department}")
#sample input
e=EmployeeID("Rahul","EMP102","IT")
e.display()
```

Problem 3: Mobile Contact Record

```python
#class to store contact details and to display them
class ContactRecord:
  def __init__(self,name,number):
    self.name=name
    self.number=number
  #function to display details
  def display(self):
    print("Contact Saved:")
    print(f"Name:{self.name}")
    print(f"Phone Number:{self.number}")
#sample input
c=ContactRecord("Anita",9876543210)
c.display()
```

Problem 4: Product Price Tag Generator

```python
#class to store product details
class product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  #function to display details
  def display(self):
    print(f"Product:{self.name}")
    print(f"Price:₹{self.price}")
#sample input
p=product("Headphones",2499)
p.display()
```

Problem 5: Movie Rating Display System

```python
# class that stores movie name and rating and displays movie details
class movie:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating
  #function to display details
  def display(self):
    print(f"Movie:{self.name}")
    print(f"Rating:{self.rating}/5")
#sample input
m=movie("Inception",4.8)
m.display()
```

Problem 6: Delivery Address Manager

```python
#class that stores customer name and delivery address and prints delivery details
class delivery:
  def __init__(self,name,address):
    self.name=name
    self.address=address
  #function to display details
  def display(self):
    print("Delivery Details:")
    print(f"Customer:{self.name}")
    print(f"Address:{self.address}")
#sample input
d=delivery("Suman","Hyderbad")
d.display()
```
